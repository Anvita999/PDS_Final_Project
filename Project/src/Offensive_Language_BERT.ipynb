```python
import pandas as pd
import numpy as np
import ktrain
from ktrain import text
import tensorflow as tf
```

```python
import io
```

```python
from google.colab import files  
uploaded = files.upload()
```

```python
data_test = pd.read_csv(io.BytesIO(uploaded['labeled_data.csv']))
data_train = pd.read_csv(io.BytesIO(uploaded['labeled_data.csv']))
```

```python
data_test=data_test.iloc[:,5:]
```

```python
data_train=data_train.iloc[:,5:]
```

```python
data_train.shape
```

```python
text.print_text_classifiers()
```

```python
(train, val, preproc) = text.texts_from_df(train_df=data_train, text_column='tweet', label_columns='class',
                   val_df = data_test,
                   maxlen = 400,
                   preprocess_mode = 'distilbert')
```

```python
model = text.text_classifier(name = 'distilbert', train_data = train, preproc=preproc)
```

```python
learner = ktrain.get_learner(model = model,
                             train_data = train,
                             val_data = val,
                             batch_size = 6)
```

```python

learner.fit_onecycle(lr = 2e-5, epochs=2)
```

```python
predictor = ktrain.get_predictor(learner.model, preproc)
```

```python
data = ['She is a bitch',
       'U are a bastard' , 'He is a nice guy']
```

```python
predictor.predict(data)
```
